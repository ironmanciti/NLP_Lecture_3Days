## 2. í…ìŠ¤íŠ¸ ì •ì œ (Text Cleaning)

```python
def clean_text_basic(text):
    """
    ê¸°ë³¸ í…ìŠ¤íŠ¸ ì •ì œ í•¨ìˆ˜
    - HTML íƒœê·¸ ì œê±°
    - URL ì œê±°
    - ì´ë©”ì¼ ì œê±°
    - ì „í™”ë²ˆí˜¸ ì œê±°
    - ì¤‘ë³µ ê³µë°± ì •ë¦¬

    Args:
        text: ì…ë ¥ í…ìŠ¤íŠ¸

    Returns:
        ì •ì œëœ í…ìŠ¤íŠ¸
    """
    # HTML íƒœê·¸ ì œê±°
    text = re.sub(r'<[^>]+>', '', text)     # >ê°€ ì•„ë‹Œ ì–´ë–¤ ë¬¸ìë¥¼ > ê°€ ë‚˜ì˜¬ ë•Œê¹Œì§€

    # URL ì œê±°
    text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text)

    # ì´ë©”ì¼ ì œê±°
    text = re.sub(r'\S+@\S+', '', text)

    # ì „í™”ë²ˆí˜¸ ì œê±° (í•œêµ­ í˜•ì‹: 02-1234-5678, 010-1234-5678 ë“±)
    text = re.sub(r'\d{2,3}-\d{3,4}-\d{4}', '', text)

    # ì´ëª¨ì§€ ë³€í™˜
    text = emoji.replace_emoji(text, replace='')

    # í•œê¸€, ì˜ë¬¸, ìˆ«ìë§Œ ìœ ì§€
    text = re.sub(r'[^ê°€-í£a-zA-Z0-9\s]', '', text)

    # ì¤‘ë³µ ê³µë°± ì •ë¦¬
    text = re.sub(r'\s+', ' ', text)

    return text.strip()

# í…ŒìŠ¤íŠ¸
print("=" * 80)
print("[ê¸°ë³¸ ì •ì œ ê²°ê³¼]")
print("=" * 80)
for i, text in enumerate(sample_texts, 1):
    cleaned = clean_text_basic(text)
    print(f"\nì›ë³¸ {i}: {text}")
    print(f"ì •ì œ {i}: {cleaned}")
```

## 3. ë¶ˆìš©ì–´ ì œê±°

```python
# í•œêµ­ì–´ ë¶ˆìš©ì–´ ë¦¬ìŠ¤íŠ¸ (ê¸°ë³¸)
korean_stopwords = [
    'ì´', 'ê°€', 'ì„', 'ë¥¼', 'ì—', 'ì˜', 'ì™€', 'ê³¼', 'ë„', 'ë¡œ', 'ìœ¼ë¡œ',
    'ì€', 'ëŠ”', 'ì—ì„œ', 'ì—ê²Œ', 'ê»˜', 'í•œí…Œ', 'ì—ê²Œì„œ', 'í•œí…Œì„œ',
    'ì´ë‹¤', 'ìˆë‹¤', 'ë˜ë‹¤', 'í•˜ë‹¤', 'ë˜', 'í•˜', 'ê²ƒ', 'ìˆ˜', 'ë“±',
    'ê·¸', 'ì €', 'ì´ê²ƒ', 'ê·¸ê²ƒ', 'ì €ê²ƒ', 'ê·¸ëŸ°', 'ì´ëŸ°', 'ì €ëŸ°',
    'ê·¸ë¦¬ê³ ', 'ê·¸ëŸ¬ë‚˜', 'í•˜ì§€ë§Œ', 'ë˜í•œ', 'ë˜', 'ê·¸ë˜ì„œ', 'ê·¸ëŸ¬ë¯€ë¡œ'
]

def remove_stopwords(text, stopwords_list):
    """
    ë¶ˆìš©ì–´ ì œê±° í•¨ìˆ˜

    Args:
        text: ì…ë ¥ í…ìŠ¤íŠ¸
        stopwords_list: ë¶ˆìš©ì–´ ë¦¬ìŠ¤íŠ¸

    Returns:
        ë¶ˆìš©ì–´ê°€ ì œê±°ëœ í…ìŠ¤íŠ¸
    """
    words = text.split()
    filtered_words = [word for word in words if word not in stopwords_list]
    return ' '.join(filtered_words)

# í…ŒìŠ¤íŠ¸
print("=" * 80)
print("[ë¶ˆìš©ì–´ ì œê±° ê²°ê³¼]")
print("=" * 80)
test_text = "ì´ ì˜í™”ëŠ” ì •ë§ ì¬ë¯¸ìˆëŠ” ì˜í™”ì…ë‹ˆë‹¤. ê·¸ë˜ì„œ ë˜ ë³´ê³  ì‹¶ìŠµë‹ˆë‹¤."
print(f"ì›ë³¸: {test_text}")
print(f"ë¶ˆìš©ì–´ ì œê±°: {remove_stopwords(test_text, korean_stopwords)}")
```

### 4. ìˆ«ì ì²˜ë¦¬

ëª©ì  - ì˜ë¯¸ë¥¼ ë³´ì¡´í•˜ë©´ì„œ ë…¸ì´ì¦ˆë¥¼ ì¤„ì´ê¸° ìœ„í•¨

ìˆ«ìëŠ” ë¬¸ì¥ ì˜ë¯¸ì— ì¤‘ìš”í•˜ì§€ ì•Šì€ ê²½ìš°ê°€ ë§ê³ , ê°’ì´ ê³„ì† ë°”ë€Œì–´ í¬ì†Œì„±(sparsity) ì„ í‚¤ìš°ë©°, ëª¨ë¸ì´ ì¼ë°˜í™”í•˜ê¸° ì–´ë µê²Œ ë§Œë“­ë‹ˆë‹¤.

```python
def normalize_numbers(text, mode='remove'):
    """
    ìˆ«ì ì²˜ë¦¬ í•¨ìˆ˜

    Args:
        text: ì…ë ¥ í…ìŠ¤íŠ¸
        mode: 'remove' (ì œê±°), 'replace' (ìˆ«ìë¡œ ë³€í™˜), 'keep' (ìœ ì§€)

    Returns:
        ì²˜ë¦¬ëœ í…ìŠ¤íŠ¸
    """
    if mode == 'remove':
        text = re.sub(r'\d+', '', text)
    elif mode == 'replace':
        text = re.sub(r'\d+', '[NUM]', text)
    elif mode == 'keep':
        return text.strip()   # ìˆ«ì í¬í•¨ ì›ë¬¸ ìœ ì§€
    else:
        return "mode error"

    return text.strip()


# í…ŒìŠ¤íŠ¸
print("=" * 80)
print("[ìˆ«ì ì²˜ë¦¬ ê²°ê³¼]")
print("=" * 80)
test_text = "2024ë…„ 3ì›” 15ì¼ì— 100ëª…ì´ ì°¸ì„í–ˆìŠµë‹ˆë‹¤"
print(f"ì›ë³¸: {test_text}")
print()
print(f"ìˆ«ì ì œê±°: {normalize_numbers(test_text, mode='remove')}")
print(f"ìˆ«ì ë³€í™˜: {normalize_numbers(test_text, mode='replace')}")
print(f"ìˆ«ì ìœ ì§€: {normalize_numbers(test_text, mode='keep')}")
```

## 5. í†µí•© ì „ì²˜ë¦¬ íŒŒì´í”„ë¼ì¸

```python
def preprocessing_pipeline(text,
                           remove_stopwords=False,
                          normalize_num='keep'):
    """
    í†µí•© ì „ì²˜ë¦¬ íŒŒì´í”„ë¼ì¸

    Args:
        text: ì›ë³¸ í…ìŠ¤íŠ¸
        remove_stopwords: ë¶ˆìš©ì–´ ì œê±° ì—¬ë¶€
        normalize_num: ìˆ«ì ì²˜ë¦¬ ë°©ì‹ ('remove', 'replace', 'keep')

    Returns:
        ì „ì²˜ë¦¬ ê²°ê³¼ ë”•ì…”ë„ˆë¦¬
    """
    # 1. ê¸°ë³¸ ì •ì œ (HTML, URL, ì´ë©”ì¼, ì „í™”ë²ˆí˜¸, ì´ëª¨ì§€, íŠ¹ìˆ˜ë¬¸ì ì œê±°)
    text = clean_text_basic(text)

    # 2. ìˆ«ì ì •ê·œí™”
    text = normalize_numbers(text, mode=normalize_num)

    # 3. ë¶ˆìš©ì–´ ì œê±° (ì„ íƒ)
    if remove_stopwords:
        text = remove_stopwords(text, korean_stopwords)

    return {'cleaned_text': text}

# í…ŒìŠ¤íŠ¸
print("=" * 80)
print("[í†µí•© ì „ì²˜ë¦¬ íŒŒì´í”„ë¼ì¸]")
print("=" * 80)

test_texts = [
    "ì•ˆë…•í•˜ì„¸ìš”! ğŸ˜Š ì˜¤ëŠ˜ ë‚ ì”¨ê°€ ì •ë§ ì¢‹ë„¤ìš”. https://example.com",
    "ì´ ì˜í™”ëŠ” 2024ë…„ì— ê°œë´‰í•œ ìµœê³ ì˜ ì˜í™”ì…ë‹ˆë‹¤!!! ğŸ‘"
]

for text in test_texts:
    print(f"\nì›ë³¸: {text}")
    result = preprocessing_pipeline(
        text,
        remove_stopwords=False,
        normalize_num='replace'
    )
    print(f"ì •ì œ: {result['cleaned_text']}")
```

```python
# ìƒ˜í”Œ ì˜í™” ë¦¬ë·° ë°ì´í„°
movie_reviews = [
    "ì´ ì˜í™” ì •ë§ ìµœê³ ì…ë‹ˆë‹¤!!! ğŸ˜ŠğŸ˜ŠğŸ˜Š ì¬ë¯¸ìˆì–´ìš”~",
    "ë³„ë¡œ ì¬ë¯¸ì—†ì–´ìš”... https://movie.naver.com ë°©ë¬¸í•´ë³´ì„¸ìš”",
    "2024ë…„ ìµœê³ ì˜ ì‘í’ˆì´ë¼ê³  ìƒê°í•©ë‹ˆë‹¤. email@example.com",
    "<div>HTML íƒœê·¸</div>ê°€ í¬í•¨ëœ ë¦¬ë·°ì…ë‹ˆë‹¤. 02-1234-5678",
    "ë„ˆë¬´ë„ˆë¬´ë„ˆë¬´   ì¢‹ì•„ìš”!!! ğŸ‘ğŸ‘"
]

print("=" * 80)
print("[ì‹¤ì „ ì˜ˆì œ: ì˜í™” ë¦¬ë·° ì „ì²˜ë¦¬]")
print("=" * 80)

# ì „ì²˜ë¦¬ íŒŒì´í”„ë¼ì¸ ì ìš©
processed_reviews = []
for review in movie_reviews:
    processed = preprocessing_pipeline(
        review,
        remove_stopwords=False,
        normalize_num='replace'
    )
    processed_reviews.append({
        'ì›ë³¸': review,
        'ì •ì œ': processed['cleaned_text']
    })

# ê²°ê³¼ í™•ì¸
df_reviews = pd.DataFrame(processed_reviews)
print("\n[ì „ì²˜ë¦¬ ê²°ê³¼]")
df_reviews
```
